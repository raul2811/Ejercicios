```python
#Grafica El Sexo De los Estudiantes
def grafica_sexo(file,nombre):#genera grafica de sexo  (datos y nombre a asignar)
    sex_F=file.loc[:,["sex"]]
    sex_F=sex_F.loc[(file!= "F").all(axis=1),:]
    sex_M=file.loc[:,["sex"]]
    sex_M=sex_M.loc[(file!= "M").all(axis=1), :]
    sex_M=(len(sex_M))
    sex_F=(len(sex_F))
    df_sex=pd.DataFrame({"cantidad": [sex_M, sex_F],},index=["M", "F"])
    grafica_sex=df_sex.plot.pie(y='cantidad', figsize=(5, 5),colors=['lightblue','lightpink'],autopct='%1.2f%%')
    plt.legend(title="Genero", loc="upper right")
    plt.axis("equal")
    plt.title("Grafica De Sexo Curso Matematicas"+nombre+"\n\n")#titulo de la grafica
    plt.savefig("Graficas\grafica_De_Sexo"+nombre+".png")#guardamos la grafica en un png
    plt.savefig("Graficas\grafica_De_Sexo"+nombre,dpi=2000, bbox_inches="tight", transparent=True)#parametros de guardato
#----------------------#
#Grafica De cantidad de estudiantes
def grafica_edades(file,nombre):#genera grafica de edades (datos y nombre a asignar)
    df_age=file.loc[:,["age"]]
    df_edades=(df_age.pivot_table(index = ["age"], aggfunc ='size'))
    colors= ['lightgrey', 'yellow', 'lightblue', 'red', 'green', 'lightpink']
    edades=list(df_edades.index)
    cantidad=list(df_age.pivot_table(index = ["age"], aggfunc ='size'))
    grafica=plt.bar(edades,cantidad,width=0.5,color=colors)
    plt.title("Grafica De cantidad de estudiantes que tienen la misma edad"+nombre)
    plt.xlabel("Edades")
    plt.ylabel("Cantidad")
    plt.savefig("Graficas\Grafica_De_cantidad_de_estudiantes_que_tienen_la_misma_edad"+nombre+".png")#guardamos la grafica en un png
    plt.savefig("Graficas\Grafica_De_cantidad_de_estudiantes_que_tienen_la_misma_edad"+nombre,dpi=2000, transparent=True)#parametros de guardato
#-----------------------#   
def promedio_edades(file,nombre):#genera Una impresion del promedio de todas las edades
    df_age=file.loc[:,["age"]]
    df_edades=(df_age.pivot_table(index = ["age"], aggfunc ='size'))
    edades=list(df_edades.index)
    promedio_edades=statistics.mean(edades)
    print("El Promedio De Edades De {} Es:{:.2f}".format(nombre,promedio_edades))
#----------------------#
def promedio_notas(file,nombre):#genera el promedio de las notas G1, G2, G3  (datos y nombre a asignar)
    df_promedios=file.loc[:,["G1","G2","G3"]]
    colors= ['yellow', 'red', 'green',]
    nombres=["G1","G2","G3"]
    G1=list(df_promedios["G1"])
    G2=list(df_promedios["G2"])
    G3=list(df_promedios["G3"])
    promedio_G1=statistics.mean(G1)
    promedio_G2=statistics.mean(G2)
    promedio_G3=statistics.mean(G3)
    promedios=[promedio_G1,promedio_G2,promedio_G3]
    print("El Promedio De G1 De {} Es:{:.2f}".format(nombre,promedio_G1))
    print("El Promedio De G2 De {} Es:{:.2f}".format(nombre,promedio_G2))
    print("El Promedio De G3 De {} Es:{:.2f}".format(nombre,promedio_G3))
    plt.barh(nombres,promedios,height=0.2,color=colors) 
    plt.ylabel("G#") 
    plt.xlabel("Calificacion")  
    plt.title("Grafica_De_promedio_de_las_notas_de_G1_G2_G3")
    plt.savefig("Graficas\Grafica_De_promedio_de_las_notas_de_G1_G2_G3"+nombre+".png")#guardamos la grafica en un png
    plt.savefig("Graficas\Grafica_De_promedio_de_las_notas_de_G1_G2_G3"+nombre,dpi=2000, transparent=True)#parametros de guardato
def promedio_notas_sin_grafica(file,nombre):#genera el promedio de las notas G1, G2, G3  (datos y nombre a asignar)
    df_promedios=file.loc[:,["G1","G2","G3"]]
    colors= ['yellow', 'red', 'green',]
    nombres=["G1","G2","G3"]
    G1=list(df_promedios["G1"])
    G2=list(df_promedios["G2"])
    G3=list(df_promedios["G3"])
    promedio_G1=statistics.mean(G1)
    promedio_G2=statistics.mean(G2)
    promedio_G3=statistics.mean(G3)
    promedios=[promedio_G1,promedio_G2,promedio_G3]
    print("El Promedio De G1 De {} Es:{:.2f}".format(nombre,promedio_G1))
    print("El Promedio De G2 De {} Es:{:.2f}".format(nombre,promedio_G2))
    print("El Promedio De G3 De {} Es:{:.2f}".format(nombre,promedio_G3))
```

```python
def ausencias_grafica(file,nombre):#genera Una impresion del promedio de todas las edades
    df_promedios=file.loc[:,["G1","G2","G3"]]
    nombres=["G1","G2","G3"]
    G1=list(df_promedios["G1"])
    G2=list(df_promedios["G2"])
    G3=list(df_promedios["G3"])
    df_absences=file.loc[:,["absences"]]
    df_ausencias=(df_absences.pivot_table(index = ["absences"], aggfunc ='size'))
    ausencias=df_absences['absences'].tolist()
    numero_clases=max(ausencias)
    cantidad=list(df_absences.pivot_table(index = ["absences"], aggfunc ='size'))
    porcentaje_de_asistencia=[]
    for x in range(len(ausencias)):
        porcentaje_de_asistencia.append(75-ausencias[x])
        df_porcentaje_de_asistencia=pd.DataFrame(porcentaje_de_asistencia,columns=["extra"])
    porcentaje_de_asistencia_approved=[]
    for i in range(len(ausencias)):
        if porcentaje_de_asistencia[i] < 60:
            porcentaje_de_asistencia_approved.append(0)
        elif porcentaje_de_asistencia[i] > 60 and G3[i]<10:
            porcentaje_de_asistencia_approved.append(0)
        elif porcentaje_de_asistencia[i] > 60 and G3[i]>=10 and G3[i]<=15 :
            porcentaje_de_asistencia_approved.append(1)
        elif porcentaje_de_asistencia[i] > 60 and G3[i]>15:
            porcentaje_de_asistencia_approved.append(1)
        else:
            porcentaje_de_asistencia_approved.append(0)
    df_porcentaje_de_asistencia_approved=pd.DataFrame(porcentaje_de_asistencia_approved,columns=["approved"])
    contador=df_porcentaje_de_asistencia_approved["approved"].value_counts()
    approved_1=contador[1]
    approved_0=contador[0]
    df_approved=pd.DataFrame({"approved": [approved_0, approved_1],},index=["Reprobado", "Aprobado"])
    grafica_approved=df_approved.plot.pie(y='approved', figsize=(5, 5),colors=['red','green'],autopct='%1.2f%%')
    plt.legend(title="Resultado", loc="upper right")
    plt.axis("equal")
    plt.title("Graficas\grafica_De_aprobados_reprobados"+nombre+"\n\n")#titulo de la grafica
    plt.savefig("Graficas\grafica_De_aprobados_reprobados"+nombre+".png")#guardamos la grafica en un png
    plt.savefig("Graficas\grafica_De_aprobados_reprobados"+nombre,dpi=2000, bbox_inches="tight", transparent=True)#parametros de guardato
    csv=pd.concat([file,df_porcentaje_de_asistencia_approved],axis=1)
    variable=("student{}_resultado_.csv".format(nombre))
    csv.to_csv(variable,sep=";")
```
